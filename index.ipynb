## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

A primary key is a column that serves as a unique identifier of a specific table. The values uniquely identify a row within that table. Foreign keys are the primary keys of other data tables. Basically a data table will have at least one primary key as well as other columns that correspond to columns with keys that serve as primary keys for those other ("foreign") data tables. Joins, a mechanism for combining information from separate data tables are performed by matching primary and foreign keys between tables. 



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Different entities within a relational database can have different types of relationships including one-to-one, one-to-many, and many-to-many relationships. 

In a one-to-one relationship, only one record (row) from each table (side of the relationship) is connected to just a single record from the other table. For example, each individual in a database has a single social security number and each social security number is only assigned to a single individual (ignoring the fact that identity theft can happen).

In a one-to-many relationship, a single record from one table may be connected to zero, one, or multiple records in the other table. For example, each individual only has one birthday, but that specific birthdate is shared among many individuals around the world.

In a many-to-many relationship, zero, one, or multiple records from one table may match with zero, one, or many records from the other table. For example, a musician or band may fit into multiple subgenres of music and each of those subgenres also encompasses many other artists.



## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Different versions of SQL perform different types of joins. SQLite does not support outer joins. Outer joins return all records from both tables even if they don't contain matching keys. Inner joins only return records with keys that match from both tables. This way avoids bringing in a bunch of null values by simply dropping records for which a key is not found in both tables. Left and right joins operate in essentially the same way and the result just differs based on which table is listed first (the left table) and which is referenced second (the right table). Essentially in a left (right) join you retain all keys and their associated records from the left (right) table and then add to those records data from the other table where the key matches a key in the first table.


## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

In order to use an aggregate function, you have to have a GROUP BY statement. The GROUP BY statement groups your results based on the unique values found in the column(s) you specify and returns 1 record per unique group. This record is a summary row that can report the results of whatever aggregate function you called. That may be the mean, min, max, count, sum, etc. of each group for another specified column.
